```python
template = """
; Tricky patterns      
trickyPattern0XPosArray
        .BYTE $08,$55  
        .BYTE $13,$55  
        .BYTE $07,$55  
        .BYTE $12,$55  
        .BYTE $02,$55  
        .BYTE $0B,$55  
        .BYTE $09,$55  
                       
trickyPattern0YPosArray
        .BYTE $FD,$55  
        .BYTE $F5,$55  
        .BYTE $06,$55  
        .BYTE $F4,$55  
        .BYTE $FC,$55  
        .BYTE $12,$55  
        .BYTE $09,$55  
"""
```

```python
possibles = [hex(i)[2:].upper().zfill(2) for i in range(230,255)] 
possibles += [hex(i)[2:].upper().zfill(2) for i in range(1,15)]
```

```python
import random
for i in range(0,7):
    xpos = '\n\t'.join([f".BYTE ${random.choice(possibles)},$55" for i in range(0,7)])
    ypos = '\n\t'.join([f".BYTE ${random.choice(possibles)},$55" for i in range(0,7)])
    print(f"trickyPattern{i}XPosArray")
    print('\t'+xpos)
    print(f"trickyPattern{i}YPosArray")
    print('\t'+ypos)
    
```
